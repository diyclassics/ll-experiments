```python
prop_raw = next(tess.texts('texts/propertius.elegies.tess'))
```

```python
prop_lines = prop_raw.split('\n')
```

```python
prop_couplets = ["\n".join(item) for item in list(zip(prop_lines[:-1], prop_lines[1:]))[::2]]
```

```python
# Repeated word > 3 len in couplet lines

counter = 0

for couplet in prop_couplets:
    line1, line2 = couplet.split('\n')
    words1 = [w for w in line1.split()]
    words2 = [w for w in line2.split()]
    for word in words1:
        for word_ in words2:
            if word==word_ and len(word) > 3:
                print(f'{word}/{word_}:\n{couplet}\n')
                counter += 1
print(f'There are {counter} lines like this in Propertius.')
```
